```python
def convert_to_num(text): 
    num_map = {
        "one": 1,
        "two": 2,
        "three": 3,
        "four": 4,
        "five": 5,
        "six": 6,
        "seven": 7,
        "eight": 8,
        "nine": 9
    }
    
    indices = []
    for num in num_map:
        if num in text:
            start = 0
            count = text.count(num)
            found = 0
            while found < count:
                val = text.index(num, start)
                start = val+1
                found += 1
                indices.append((val, num))
            
    indices.sort()
    text_list = list(text)
    for idx, num in indices:
        text_list[idx] = str(num_map[num])
        
    return "".join(text_list)  
```
