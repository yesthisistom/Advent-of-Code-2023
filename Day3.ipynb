```python
def is_symbol(value):
    return not value.isnumeric() and value != "."
        
```

```python
def touches_symbol(table, row_val, col_min, col_max):
    for row in range(row_val-1, row_val+2):
        if row < 0 or row >= len(table):
            continue
        
        for col in range(col_min - 1, col_max + 2):
            if col < 0 or col >= len(table[0]):
                continue
                
            # If symbol, return true
            if is_symbol(table[row][col]):
                return True
            
    return False
```

```python
def create_number(row, start_idx, end_idx):
    return int("".join(row[start_idx:end_idx+1]))
```

```python
def identify_numbers(row_idx, full_table):
    row = full_table[row_idx]
    num_list = []
    
    num_start = -1
    for idx, val in enumerate(row):
        if val.isnumeric():
            # If we already have a start, continue
            if num_start == -1:
                num_start = idx
        else:
            # If we have a start, find out if it touches symbols
            if num_start != -1:
                if touches_symbol(full_table, row_idx, num_start, idx-1):
                    num_list.append(create_number(row, num_start, idx-1))
                
                num_start = -1
            
    # Check if we have a start
    if num_start != -1:
        if touches_symbol(full_table, row_idx, num_start, len(row)-1):
            num_list.append(create_number(row, num_start, len(row)-1))
            
    return num_list
    
```

```python
def get_full_number(table, num_row, num_col):
    index_list = []
    row = table[num_row]
    
    # search forwards and backwards for the full number
    for col_idx in range(num_col-1, -1, -1):
        if not row[col_idx].isnumeric():
            break
        index_list.append((num_row, col_idx))
    
    for col_idx in range(num_col, len(row)):
        if not row[col_idx].isnumeric():
            break
            
        index_list.append((num_row, col_idx))
        
    # Sort the list, create the number string
    index_list.sort()
    
    number = int("".join([row[col] for _, col in index_list]))
    
    # Returns the number, AND the index
    return number, index_list
```

```python
def identify_numbers(table, gear_row, gear_col):
    numbers = []
    numbers_indices = [] # tuples of row, col
    
    for row in range(gear_row-1, gear_row+2):
        if row < 0 or row >= len(table):
            continue
        
        for col in range(gear_col - 1, gear_col + 2):
            if col < 0 or col >= len(table[0]):
                continue
                
            # If symbol, return true
            if table[row][col].isnumeric():
                # CHECK THE INDEX TABLE
                if (row, col) not in numbers_indices:
                    number, indices = get_full_number(table, row, col)
                    numbers.append(number)
                    numbers_indices.extend(indices)
    
    return numbers
                    
                
```
